## üìñ Background
You're applying for a summer internship at a national museum for natural history. The museum recently created a database containing all dinosaur records of past field campaigns. Your job is to dive into the fossil records to find some interesting insights, and advise the museum on the quality of the data. 

## üíæ The data

### You have access to a real dataset containing dinosaur records from the Paleobiology Database ([source](https://paleobiodb.org/#/)):


| Column name | Description |
|---|---|
| occurence_no | The original occurrence number from the Paleobiology Database. |
| name | The accepted name of the dinosaur (usually the genus name, or the name of the footprint/egg fossil). |
| diet | The main diet (omnivorous, carnivorous, herbivorous). |
| type | The dinosaur type (small theropod, large theropod, sauropod, ornithopod, ceratopsian, armored dinosaur). |
| length_m | The maximum length, from head to tail, in meters. |
| max_ma | The age in which the first fossil records of the dinosaur where found, in million years. |
| min_ma | The age in which the last fossil records of the dinosaur where found, in million years. |
| region | The current region where the fossil record was found. |
| lng | The longitude where the fossil record was found. |
| lat | The latitude where the fossil record was found. |
| class | The taxonomical class of the dinosaur (Saurischia or Ornithischia). |
| family | The taxonomical family of the dinosaur (if known). |

The data was enriched with data from Wikipedia.

```r
# Import the tidyverse library
library(tidyverse)
# Load the data
dinosaurs <- read_csv('data/dinosaurs.csv', show_col_types = FALSE)
```

```r
# Preview the dataframe
dinosaurs
```

## 1. How many different dinosaur names are present in the data?

```r
#lets check the number of unique dinosaur names
length(unique(dinosaurs$name))
```

## 2. Which was the largest dinosaur? What about missing data in the dataset?

### Missing Data

```r
#lets check on the missing data in the dataset
NAs =colSums(is.na(dinosaurs))
names(dinosaurs)[NAs>0]
```

**The columns of diet, type, length_m, region and family have NAs**

```r
missing_values_count <- colSums(is.na(dinosaurs))

# Print the result
print(missing_values_count)
```

**This tells us that there are 1355 missing in the diet column, 42 missing in the region column, 1355 missing in the type column, and 1383 missing in the length_m column. For the family, since some might be unknown, we can say that there are 1457 dinosaurs with a missing or unknown family.**

### Largest Dinosaur

```r
#lets look at the first part on the largest dinosaur
largestdinosaur= dinosaurs %>% 
select(name,length_m) %>% 
group_by(name) %>% 
summarise(largest = max(length_m,  na.rm = FALSE)) %>% 
arrange(desc(largest)) %>% 
head(5)

print(largestdinosaur)
```

**Hence, we can see that Argentinosaurus and Supersaurus were the largest dinosaurs, with a maximum length from head to tail of 35 meters.**

## 3. What dinosaur type has the most occurrences in this dataset? Create a visualization (table, bar chart, or equivalent) to display the number of dinosaurs per type. Use the AI assistant to tweak your visualization (colors, labels, title...).

```r
#lets look for the most occuring dinosaur type
mostoccured = dinosaurs %>%
select(type) %>%
filter(!is.na(type)) %>%
group_by(type) %>%
summarise(No_of_dinosaurs = n()) %>%
arrange(desc(No_of_dinosaurs))

print(mostoccured)
```

**Hence, from the table above, we can see that the most common dinosaur type is the Ornithopod, with 811 dinosaurs.**

```r
# Create a bar chart to display the number of dinosaurs per type
ggplot(mostoccured, aes(x = reorder(type, -No_of_dinosaurs), y = No_of_dinosaurs, fill = type)) +
  geom_bar(stat = "identity") +
  labs(title = "Number of Dinosaurs per Type",
       x = "Dinosaur Type",
       y = "Number of Dinosaurs") +
  theme_minimal() +
  theme(axis.text.x = element_text(angle = 45, hjust = 1),
        legend.position = "none",
        plot.title = element_text(hjust = 0.5, size = 16, face = "bold"),
        axis.title.x = element_text(size = 14, face = "bold"),
        axis.title.y = element_text(size = 14, face = "bold"),
        axis.text = element_text(size = 12)) +
scale_fill_manual(values=c("blue", "red","green", "orange","purple","black"))
```

## 4. Did dinosaurs get bigger over time? Show the relation between the dinosaur length and their age to illustrate this.

```r
#relation between the dinosaur length and age
dinosaurs_length_with_time <- dinosaurs %>%
  select(length_m, max_ma, min_ma) %>% 
filter(!is.na(length_m)) %>% 
  group_by(length_m) %>%
  reframe(age = floor(max_ma - min_ma))

#preview
print(dinosaurs_length_with_time)
```

```r
# Create a scatter plot to show the relationship between dinosaur length and their age
ggplot(dinosaurs_length_with_time, aes(x = age, y = length_m)) +
  geom_point(color = "blue", alpha = 0.6) +
  geom_smooth(method = "lm", formula = y ~ x, color = "red", se = FALSE) +
  labs(title = "Relationship Between Dinosaur Length and Age",
       x = "Age (millions of years)",
       y = "Length (meters)") +
  theme_minimal() +
  theme(plot.title = element_text(hjust = 0.5, size = 16, face = "bold"),
        axis.title.x = element_text(size = 14, face = "bold"),
        axis.title.y = element_text(size = 14, face = "bold"),
        axis.text = element_text(size = 12))
```

**We can see from the scatter plot that the lower the length the longer the dinosar lives**

## 5. Use the AI assitant to create an interactive map showing each record.

```r
# Load necessary libraries
library(leaflet)
library(dplyr)

# Prepare the data for the map
dinosaurs_map_data <- dinosaurs %>%
  filter(!is.na(lat) & !is.na(lng))

# Create an interactive map with marker clusters
m <- leaflet(dinosaurs_map_data) %>%
  addTiles() %>%
  addMarkers(
    clusterOptions = markerClusterOptions(),
    ~lng, ~lat,
    popup = ~paste("<strong>Name:</strong>", name, "<br>",
                   "<strong>Diet:</strong>", diet, "<br>",
                   "<strong>Type:</strong>", type, "<br>",
                   "<strong>Length (m):</strong>", length_m, "<br>",
                   "<strong>Max Age (Ma):</strong>", max_ma, "<br>",
                   "<strong>Min Age (Ma):</strong>", min_ma, "<br>",
                   "<strong>Region:</strong>", region, "<br>",
                   "<strong>Class:</strong>", class, "<br>",
                   "<strong>Family:</strong>", family)
  )

# Display the map
m
```

## 6. Any other insights you found during your analysis?

### Classes

```r
#Looking which taxonomical class dominates
dinosaurs_taxonomy=dinosaurs %>% 
select(class) %>% 
group_by(class) %>% 
summarise(Count= n())

#lets create a pie chart to see
colors= c("red", "green")
pie(dinosaurs_taxonomy$Count, 
	labels = paste(dinosaurs_taxonomy$class, " (", dinosaurs_taxonomy$Count, ")", sep = ""),
	col =colors,
	main = "Distribution of Dinosaur Classes")

# Add a legend
legend(
  "topleft",
  legend = dinosaurs_taxonomy$class,
  fill = colors,
  title = "Dinosaur Classes"
)
```

**From the pie Chart we can see that Saurishia Class Dominates**

### Family

```r
#Investigating which dinosaur family are known
dinosaurs_family =dinosaurs %>% 
select(family) %>% 
group_by(family) %>% 
summarise(Count=n()) %>% 
arrange(desc(Count))

print(dinosaurs_family)
```

**Hence it can be seen that 1457 dinosaurs have unknown families.**

### Diet

```r
#Investigating the most common diet while removing NAs in the diet column
dinosaurs_diet =dinosaurs %>% 
select(diet) %>% 
filter(!is.na(diet)) %>% 
group_by(diet) %>% 
summarise(Total_No_of_Dinosaurs=n())

print(dinosaurs_diet)
```

```r
#A plot to show the diet distribution
ggplot(dinosaurs_diet, aes(x= diet, y =Total_No_of_Dinosaurs, fill= diet))+
geom_bar(stat="identity")+
labs(
x="Diet",
y= "Total No of Dinosaurs",
title= "Distribution of diet among the dinosaurs")+
theme_minimal()+
scale_fill_manual(values = c("blue","green","red"))+
theme(plot.title = element_text(hjust=0.5))
```

### diet versus type

```r
#looking at the relationship between type and diet
diet_type_dinosaurs= dinosaurs %>% 
select(diet, type) %>% 
filter(!is.na(diet)) %>% 
filter(!is.na(type)) %>% 
group_by(type, diet) %>%
summarise(Total_No_Of_Dinosaurs=n(), .groups = "drop")
```

```r
ggplot(diet_type_dinosaurs, aes(x = type, y = Total_No_Of_Dinosaurs, fill = diet)) +
  geom_bar(stat = "identity", position = "stack") +
  labs(title = "Distribution of Dinosaur Diets by Type",
       x = "Type",
       y = "Total Number of Dinosaurs") +
  theme_minimal() +
  theme(plot.title = element_text(hjust = 0.5, size = 16, face = "bold"),
        axis.title.x = element_text(size = 14, face = "bold"),
        axis.title.y = element_text(size = 14, face = "bold"),
        axis.text = element_text(size = 12),
		axis.text.x = element_text(size = 12, angle = 45, hjust = 1),
        legend.title = element_text(size = 14, face = "bold"),
        legend.text = element_text(size = 12)) +
  scale_fill_manual(values= c("red", "green", "pink"))
```

**From the bar chart, Ornithopod dinosaurs have the highest number of dinosaurs with a diet of herbivorous**

### diet versus length

```r
# Filter out rows with NA values in diet or length_m
dinosaurs_diet_length <- dinosaurs %>% 
  select(diet, length_m) %>% 
  filter(!is.na(diet) & !is.na(length_m))

# Create the box plot to show the relationship between diet and length
ggplot(dinosaurs_diet_length, aes(x = diet, y = length_m, fill = diet)) +
  geom_boxplot() +
  labs(
    title = "Relation Between Diet and Length",
    x = "Diet",
    y = "Length (meters)"
  ) +
  theme_minimal() +
  theme(
    plot.title = element_text(hjust = 0.5, size = 16, face = "bold"),
    axis.title.x = element_text(size = 14, face = "bold", angle = 45, hjust = 1),
    axis.title.y = element_text(size = 14, face = "bold"),
    axis.text.x = element_text(size = 12, angle = 45, hjust = 1),
    axis.text.y = element_text(size = 12)
  )
```

You have just been hired by a large movie studio to perform data analysis. Your manager, an executive at the company, wants to make new movies that "recapture the magic of old Hollywood." ## üíæ The data

### You have access to the following table, _**cinema.films**_:

| Column name | Description |
|---|---|
| id | Unique movie identifier. |
| title | The title of the movie. |
| release_year | The year the movie was released to the public. |
| country | The country in which the movie was released. |
| duration | The runtime of the movie, in minutes. |

## 2. There seems to be a lot of missing data in the gross and budget columns. How many rows have missing data? What would you recommend your manager to do with these rows?

```r
SELECT COUNT(*) AS num_missing
FROM cinema.films
WHERE gross IS NULL
   OR budget IS NULL;
```

## 3. How many different certifications or ratings are present in the database?

```r
SELECT count(distinct(certification))
FROM cinema.films;
```

**There are 13 different certification or ratings present in the database.**

## 4. What are the top five countries in terms of number of movies produced?

```r
-- top five countries in movie production
SELECT count(title) AS Total_No_of_movies, country
FROM cinema.films
GROUP BY country
ORDER BY Total_No_of_movies DESC
LIMIT 5;
```

**The top five countries in production are USA, UK, France, Canada and Germany.**

## 5. What is the average duration of English versus French movies? (Don't forget you can use the AI assistant!)

```r
-- English Versus French
SELECT language, floor(AVG(duration)) AS average_duration
FROM cinema.films
WHERE language IN ('English', 'French')
GROUP BY language;
```

**Hence, the average duration for English movies is 107 minutes, while French movies are 104 minutes, which means English movies have a duration of 3 minutes more.**

## 6. Any other insights you found during your analysis?

### Ratings

```r
-- checking the certification or rating that was given to most movies
SELECT certification, COUNT(certification) AS total_certifications
FROM cinema.films
GROUP BY certification
ORDER BY total_certifications DESC
LIMIT 5;
```

**2118 movies have been given the R certification or rating which is the highest number of certifications given to most movies.**

### Language Versus Ratings

```r
SELECT certification, language, count(*) AS Total_No_of_Movies
FROM cinema.films
GROUP BY cinema.films.certification, cinema.films.language
ORDER BY Total_No_of_Movies DESC;
```

**The best movies produced with an R rating were in English amounting to 1995 movies**

### Ratings versus Years

```r
SELECT 
    certification, 
    CASE 
        WHEN release_year >= 1900 AND release_year < 1950 THEN '1900-1950'
        WHEN release_year >= 1930 AND release_year < 2000 THEN '1951-2000'
        WHEN release_year >= 1960 AND release_year < 2020 THEN '2001-2020'
        ELSE 'Other'
    END AS release_year_range,
    COUNT(*) AS film_count
FROM 
    cinema.films
WHERE 
    certification = 'R'
GROUP BY 
    certification, release_year_range
ORDER BY 
    release_year_range;
```

**Most movies produced during the 20th century received the most R ratings compared to movies produced during the 1990s.**
